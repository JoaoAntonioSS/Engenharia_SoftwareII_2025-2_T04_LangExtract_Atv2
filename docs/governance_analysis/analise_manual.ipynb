# Relatório de Análise Manual: Governança do Projeto
## Combinação de Estratégia de Releases e Fluxo de Trabalho

**Projeto Analisado:** LangExtract  
**Responsável:** Isis Gabrielle  
**Metodologia:** Inspeção manual do repositório GitHub, considerando a estratégia de releases, estrutura de branches, histórico de commits e uso de pull requests.


## 1. Modelo de Fluxo de Trabalho Identificado

**Modelo:** Branch principal única (`main`) com branches temporárias para Pull Requests  
*(Fluxo compatível com GitHub Flow / Trunk-Based Development simplificado)*

**Evidências:**  
A análise da estrutura de branches do repositório indica a existência de múltiplas branches (49 no total). Entretanto, apenas a branch `main` é mantida como branch permanente e central do desenvolvimento. As demais branches são utilizadas de forma temporária, majoritariamente associadas à implementação de funcionalidades, correções ou experimentações vinculadas a Pull Requests.

Não foram identificadas branches persistentes com papéis claramente definidos, como `develop`, `release/*` ou `hotfix/*`, o que indica a adoção de um fluxo de trabalho simplificado e centralizado no branch principal.


## 2. Consistência entre Releases e Fluxo de Trabalho

**Consistência Identificada:** Alinhamento entre a branch principal e a geração de releases

**Evidências:**  
As releases publicadas no projeto estão associadas a estados estáveis do código presentes na branch `main`, indicando que apenas versões consolidadas do software são utilizadas como base para novos lançamentos. O versionamento ocorre após a integração das mudanças por meio de Pull Requests, o que demonstra coerência entre o fluxo de desenvolvimento e a estratégia de releases.

Além disso, o uso de versionamento semântico e tags reforça que os lançamentos são realizados de forma controlada, evitando a criação de releases a partir de branches temporárias ou experimentais.


## 3. Indícios de Governança do Projeto

### Indícios Observados
- Utilização consistente de versionamento semântico (SemVer) nas releases.  
- Presença de Pull Requests como mecanismo de controle, revisão e integração de código.  
- Existência de múltiplas branches temporárias para desenvolvimento de funcionalidades, indicando colaboração ativa.  
- Histórico de commits organizado, com mensagens que descrevem claramente o tipo de alteração realizada.

### Limitações Identificadas
- Ausência de um modelo de branching formalmente documentado.  
- Inexistência de branches persistentes com papéis específicos no fluxo de desenvolvimento.  
- Documentação limitada sobre práticas de governança e processo de desenvolvimento.


## 4. Classificação da Governança

**Classificação:** Governança **Intermediária**

**Justificativa:**  
O projeto LangExtract apresenta práticas consolidadas de controle de versões e integração de código, como releases bem definidas, uso de versionamento semântico e adoção de Pull Requests. Esses fatores indicam que o processo de desenvolvimento não é ad-hoc.

Entretanto, a ausência de um fluxo de branching mais formalizado, bem como a falta de documentação explícita sobre a estratégia de governança, impedem a classificação do projeto como possuidor de uma governança totalmente estruturada. Dessa forma, o projeto é classificado como possuidor de governança intermediária, apresentando organização e controle, mas ainda com espaço para maior formalização.
