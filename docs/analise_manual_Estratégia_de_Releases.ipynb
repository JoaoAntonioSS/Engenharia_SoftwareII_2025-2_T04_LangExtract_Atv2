# Relatório de Análise Manual: Estratégia de Releases

**Projeto Analisado:** LangExtract
**Responsável:** Débora Gonçalves
**Metodologia:** Inspeção manual do histórico de versões, tags e estrutura de branches.



## 1. Padrão de Versionamento Identificado
**Padrão:** Semantic Versioning (SemVer)

**Evidências:**
A análise do histórico de commits revela uma adesão estrita ao formato `MAJOR.MINOR.PATCH`:
* **Patch Version:** Observou-se o commit `chore: Bump version to 1.1.1` em 27 de novembro, sucedendo a versão `1.1.0`. Isso indica correções de bugs retrocompatíveis.
* **Minor Version:** Observou-se a evolução da versão `1.0.9` (31 de ago) para `1.1.0` (14 de nov), coincidindo com a adição de novas funcionalidades (ex: `feat: Add Vertex AI Batch API`), o que valida a regra de incremento *Minor*.

## 2. Estratégia de Releases
**Estratégia:** Feature-based Releases (Lançamentos baseados em funcionalidades) / Continuous Delivery

**Evidências:**
* **Ausência de Ciclos Fixos (Release Train):** As datas de lançamento são irregulares (agosto, novembro), ocorrendo apenas quando um conjunto de funcionalidades está maduro ou correções são necessárias, e não em datas de calendário fixo.
* **Fluxo Linear:** O histórico mostra um desenvolvimento contínuo na branch `main`.
* **Conventional Commits:** O uso padronizado de prefixos nos commits (`feat:`, `fix:`, `chore:`, `docs:`) sugere um processo preparado para automação de *changelogs* e versionamento, típico de fluxos de *Continuous Delivery*.

## 3. Análise de LTS (Long-Term Support)
**Status:** Não possui versão LTS.

**Evidências:**
* **Estrutura de Branches:** A inspeção da aba "Branches" mostra apenas a branch `main` como padrão e diversas branches de curta duração (`feat/...`, `fix/...`, `chore/...`).
* **Ausência de Branches de Manutenção:** Não foram encontradas branches de suporte a versões legadas (como `release/1.0` ou `support/v1`).
* **Modelo "Rolling Forward":** O desenvolvimento segue sempre para frente. Correções são aplicadas na versão atual (`main`) gerando uma nova versão imediata, sem *backporting* para versões antigas.
